```python
import numpy as np
from statistics import mean

n = np.arange(101)
x = np.cos(0.1*np.pi*n)

y = np.array(x)

y1 = y.sum()       # Soma dos elementos do array
y2 = y.max()       # Valor máximo dos elementos do array
y3 = y.min()       # Valor mínimo dos elementos do array 
y4 = mean(y)       # Valor médio dos elementos do array
y5 = np.prod(y)    # Produto dos elementos do array

r = np.array([y1,y2,y3,y4,y5])
print('Array resultado =',r)
```

Gráficos

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(101)
y = np.cos(0.1*np.pi*n)
y1 = np.exp(-.05*n)

plt.figure(figsize=(8,5))
plt.plot(n*.01,y,'b',label='Sinal cossenoidal')
plt.plot(n*.01,y1,'--k',label='Sinal aleatório')
plt.plot(n*.01,y+y1,'r',linewidth=3,label='Soma dos sinais')
plt.legend()
plt.axis([0,1,-1.1,1.1])
plt.xticks(np.arange(0,1.1,.1))
plt.yticks(np.arange(-1,2.1,.5))
plt.title('Gráfico de Sinais')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid()
plt.show()
```
