```python
import control.matlab as ctr
import matplotlib.pyplot as plt
```

```python
num1 = [0.01758,0.01540008]
den1 = [1,-1.06703,0.6703]
dt=0.2
G1 = ctr.tf(num1,den1,dt)
print(G1)
```

```python
rlist,klist=ctr.rlocus(G1, PrintGain=True, grid=True)

plt.xlim(-5,1)
plt.ylim(-1,1)
```

```python
polos = ctr.pole(G1)
zeros = ctr.zero(G1)
print('Polos =>', polos)
print('Zeros =>', zeros)
```
